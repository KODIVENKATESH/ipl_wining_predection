```python
data=pd.read_csv("all_season_batting_card.csv")
```

```python
data.head()
```

```python
data.head(20)
```

```python
data.columns
```

```python

data=data.drop(['match_id','match_name','home_team','away_team','venue','city','country','current_innings','innings_id','name','minutes','captain','runningScore','runningOver','shortText','commentary','link','strikeRate'],axis=1)
```

```python
names=data['fullName'].unique()
```

```python
runs2021=[]
ballsfaced2021=[]
fours2021=[]
sixes2021=[]
strikerate2021=[]

runs2020=[]
ballsfaced2020=[]
fours2020=[]
sixes2020=[]
strikerate2020=[]

runs2019=[]
ballsfaced2019=[]
fours2019=[]
sixes2019=[]
strikerate2019=[]

runs2018=[]
ballsfaced2018=[]
fours2018=[]
sixes2018=[]
strikerate2018=[]

runs2017=[]
ballsfaced2017=[]
fours2017=[]
sixes2017=[]
strikerate2017=[]

for i in range(len(names)):
    temp1=0
    temp2=0
    temp3=0
    temp4=0
    temp5=0
    for ri,rd in data.iterrows():
        if(rd['fullName']==names[i])and(rd['season']==2021):
            temp1+=rd['runs']
            temp2+=rd['ballsFaced']
            temp3+=rd['fours']
            temp4+=rd['sixes']
    runs2021.append(temp1)
    ballsfaced2021.append(temp2)
    fours2021.append(temp3)
    sixes2021.append(temp4)
    if(temp2==0):
        strikerate2021.append(0)
    else:
        strikerate2021.append(temp1*100/temp2)

for i in range(len(names)):
    temp1=0
    temp2=0
    temp3=0
    temp4=0
    temp5=0
    for ri,rd in data.iterrows():
        if(rd['fullName']==names[i])and(rd['season']==2020):
            temp1+=rd['runs']
            temp2+=rd['ballsFaced']
            temp3+=rd['fours']
            temp4+=rd['sixes']
    runs2020.append(temp1)
    ballsfaced2020.append(temp2)
    fours2020.append(temp3)
    sixes2020.append(temp4)
    if(temp2==0):
        strikerate2020.append(0)
    else:
        strikerate2020.append(temp1*100/temp2)

for i in range(len(names)):
    temp1=0
    temp2=0
    temp3=0
    temp4=0
    temp5=0
    for ri,rd in data.iterrows():
        if(rd['fullName']==names[i])and(rd['season']==2019):
            temp1+=rd['runs']
            temp2+=rd['ballsFaced']
            temp3+=rd['fours']
            temp4+=rd['sixes']
    runs2019.append(temp1)
    ballsfaced2019.append(temp2)
    fours2019.append(temp3)
    sixes2019.append(temp4)
    if(temp2==0):
        strikerate2019.append(0)
    else:
        strikerate2019.append(temp1*100/temp2)
        

for i in range(len(names)):
    temp1=0
    temp2=0
    temp3=0
    temp4=0
    temp5=0
    for ri,rd in data.iterrows():
        if(rd['fullName']==names[i])and(rd['season']==2018):
            temp1+=rd['runs']
            temp2+=rd['ballsFaced']
            temp3+=rd['fours']
            temp4+=rd['sixes']
    runs2018.append(temp1)
    ballsfaced2018.append(temp2)
    fours2018.append(temp3)
    sixes2018.append(temp4)
    if(temp2==0):
        strikerate2018.append(0)
    else:
        strikerate2018.append(temp1*100/temp2)


for i in range(len(names)):
    temp1=0
    temp2=0
    temp3=0
    temp4=0
    temp5=0
    for ri,rd in data.iterrows():
        if(rd['fullName']==names[i])and(rd['season']==2017):
            temp1+=rd['runs']
            temp2+=rd['ballsFaced']
            temp3+=rd['fours']
            temp4+=rd['sixes']
    runs2017.append(temp1)
    ballsfaced2017.append(temp2)
    fours2017.append(temp3)
    sixes2017.append(temp4)
    if(temp2==0):
        strikerate2017.append(0)
    else:
        strikerate2017.append(temp1*100/temp2)
```

```python
d={"Name":names,"runs":runs2021,"balls":ballsfaced2021, "fours":fours2021,"sixes":sixes2021,"sr":strikerate2021}
data2021=pd.DataFrame(d)
data2021
```
